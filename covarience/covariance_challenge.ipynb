```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
df = pd.read_csv('employee.csv')
df.head()
    
```

```python
df.set_index('Age', inplace=True)
df.head()
```

```python
a = df[]
def covariance(a,b):
    if len(a) != len(b):
        return

    a_mean = np.mean(a)
    b_mean = np.mean(b)

    sum = 0

    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))

    return sum/(len(a)-1)

```
